```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
datadir = 'plotNNBotER/'
```

```python
filenames=os.listdir(datadir)
filenames
```

```python
df = pd.read_csv(datadir + filenames[2] ,  index_col=False, skiprows=1, header = None, delimiter = "\t\t")
df1 =  pd.read_csv(datadir + filenames[3] ,  index_col=False, skiprows=1, header = None, delimiter = "\t\t")
df2 =  pd.read_csv(datadir + filenames[1] ,  index_col=False, skiprows=1, header = None, delimiter = "\t\t")
```

```python
rate = df[1000:1002]
df = df[0:1000]
rate1 = df1[1000:1002]
df1 = df1[0:1000]
rate2 = df2[1000:1002]
df2 = df2[0:1000]
```

```python
fig = plt.figure(figsize = (25,12))
plt.title("Part 2(a) - Measure of progress " + " while Learning " + list(rate1[0].iloc[[0]])[0]+ " after Learning " + list(rate[0].iloc[[0]])[0],fontsize=24  )
plt.tick_params(axis='x', labelsize='small',labelrotation=90,pad=1)
plt.tick_params(axis='y', labelsize='large',pad=1)
plt.xlabel('Rounds',fontsize=18)
plt.ylabel('Win Rate (%)',fontsize=18)
ax = plt.subplot(111)
ax.plot(df[0],100*df[1], label = 'AfterTraining')
ax.plot(df1[0],100*df1[1], label = 'WhileTraining')
ax.plot(df2[0],100*df2[1], label = 'Baseline')
max_xticks = 100
xloc = plt.MaxNLocator(max_xticks)
ax.xaxis.set_major_locator(xloc)
ax.legend( prop={'size': 30})
plt.savefig("Part2-a.png")
plt.show()
```

```python
data='plot_sarsa/'
filenames=os.listdir(data)
learned_df = pd.read_csv(data + filenames[0] ,  index_col=False, skiprows=1, header = None, delimiter = "\t\t")
learning_df =  pd.read_csv(data + filenames[1] ,  index_col=False, skiprows=1, header = None, delimiter = "\t\t")
```

```python
rate_learned = learned_df[1000:1002]
learned_df_=learned_df[0:1000]
rate_learning = learning_df[1000:1002]
learning_df_ = learning_df[0:1000]
```

```python
fig = plt.figure(figsize = (25,12))
plt.title("Part 2(b) - | On-Policy vs Off- Policy |" + " On-Policy: " + list(rate_learned[0].iloc[[0]])[0]+ " Off-Policy: " + list(rate[0].iloc[[0]])[0],fontsize=24)
plt.tick_params(axis='x', labelsize='small',labelrotation=90,pad=1)
plt.tick_params(axis='y', labelsize='large',pad=1)
plt.xlabel('Rounds',fontsize=18)
plt.ylabel('Win Rate (%)',fontsize=18)
ax = plt.subplot(111)
ax.plot(learned_df_[0],100*learned_df_[1], label = 'On-Policy')
ax.plot(df[0],100*df[1], label = 'Off-Policy')
max_xticks = 100
xloc = plt.MaxNLocator(max_xticks)
ax.xaxis.set_major_locator(xloc)
ax.legend( prop={'size': 30})
plt.savefig("Part2-b_1.png")
plt.show()
          
          
```
